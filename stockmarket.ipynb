```python
import pandas as pd
tsla = pd.read_csv("TSLA.csv",index_col="Date",parse_dates=["Date"],usecols=["Date","Adj Close"])
tsla.head()
```

```python
import yfinance as yf

ticker = "TSLA"
tesla_2023 = yf.download(ticker,start="2023-01-01",end="2023-11-17")
print(tesla_2023)
tesla_2023.info()

ibm_2023 = yf.download("IBM",start="2023-01-01",end="2023-11-17")
print(ibm_2023)

ibm_ytd = yf.download("IBM", period="ytd")
print(ibm_ytd)
```

```python
# Get intra-day data
ticker = "GME"
gme = yf.download(ticker,period="1mo",interval="1h")
print(gme)

gme2 = yf.download(ticker,period="1mo",interval="30m")
print(gme2)

gme3 = yf.download(ticker,period="1d",interval="1m")
gme3.head()
gme3.tail()

gme.describe()
```
